```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
X_test = pd.read_csv('./Datasets/test.csv', delimiter=',')
X_train = pd.read_csv('./Datasets/train.csv', delimiter=',')
y_train = pd.read_csv('./Datasets/y_train.csv', delimiter=',', header=None)
y_test = pd.read_csv('./Datasets/y_test.csv', delimiter=',', header=None)
```

### Decision Trees

```python
dt = DecisionTreeRegressor()  
dt.fit(X_train, y_train)
```

```python
y_pred = dt.predict(X_test)
y_pred = y_pred.reshape(-1,1)
```

```python
X=np.arange(X_test.shape[0])
df=pd.DataFrame(np.hstack((y_test,y_pred)))
df=df.sort_values(1)
plt.scatter(X, df[0], color="black")
plt.plot(X, df[1], color="blue", linewidth=3)
```

#### Model evaluation

```python
mean_absolute_error(y_test,y_pred)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
r2_score(y_test,y_pred)
```

### Random Forest

```python
rf = RandomForestRegressor(n_estimators=100, max_depth=10)
rf.fit(X_train, y_train.values.ravel())
```

```python
y_pred = rf.predict(X_test)
y_pred = y_pred.reshape(-1,1)
```

```python
df=pd.DataFrame(np.hstack((y_test,y_pred)))
df=df.sort_values(1)
plt.scatter(X, df[0], color="black")
plt.plot(X, df[1], color="blue", linewidth=3)
```

#### Model evaluation

```python
mean_absolute_error(y_test,y_pred)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
r2_score(y_test,y_pred)
```
